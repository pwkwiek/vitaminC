### Example of results of adapted histogram matching for one image of a solution

```python
import matplotlib.pyplot as plt
import cv2
from skimage import exposure

image = cv2.imread("E:/project_final/bulb/split_final_org/test/150/89.jpg")
matched = cv2.imread("E:/project_final/bulb/split_final_const/test/150/89.jpg")

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 6),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2):
    aa.set_axis_off()

ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
#ax1.set_title('source')
ax2.imshow(cv2.cvtColor(matched, cv2.COLOR_BGR2RGB))
#ax2.set_title('matched')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(8, 6))


for i, img in enumerate((image, matched)):
    for c, c_color in enumerate(('red', 'green', 'blue')):
        img_hist, bins = exposure.histogram(img[..., c], source_range='dtype')
        axes[i, c].plot(bins, img_hist / img_hist.max())
        img_cdf, bins = exposure.cumulative_distribution(img[..., c])
        axes[i, c].plot(bins, img_cdf)

axes[0, 0].set_title('red')
axes[0, 1].set_title('green')
axes[0, 2].set_title('blue')

plt.tight_layout()
plt.show()
```
